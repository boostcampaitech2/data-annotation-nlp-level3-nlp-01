```python
entities = {}
```

```python
from tqdm import tqdm

neg_list = []

for sentence in tqdm(다_sentences):
  if(len(sentence) > 512): continue
  ner_list = ner(sentence)
  sub_obj = 0
  for _, entity in ner_list:
    if entity not in entities.keys():
      entities[entity] = 1
    elif entity in ['ORGANIZATION', 'PERSON']:
      sub_obj += 1
      entities[entity] += 1
    else:
      entities[entity] += 1
  if sub_obj < 2:
    neg_list.append(sentence)
```
